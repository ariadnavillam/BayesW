```python
import os
import itertools
import random
from math import log, exp, sqrt
from decimal import *
import pandas as pd

from IPython.display import clear_output

import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib.colors import LinearSegmentedColormap
from cycler import cycler


import matplotlib.animation as animation

import numpy as np
from scipy import stats
from numpy.linalg import norm
```

```python
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
rc('font',**{'family':'serif','serif':['Palatino'], 'size':14})
rc('text', usetex=True)
plt.rc('axes', prop_cycle=(cycler('color', ['red', 'gray', 'black', 'blue', 'green'])))#, )))
```

```python
df = pd.read_csv("/run/user/1005944/gvfs/smb-share:server=fs.ista.ac.at,share=home/avillanu/hydra/example/out_bayesW/weibull_1000_200.csv", header=None)
num_mix = df[7][0]
header = ["iter", "mu", "sigmaG", "alpha", "h2", "num_markers", "num_groups", "num_mixtures"]
header.extend(["sigmaG_"+str(i) for i in range(num_mix)])
header.append(str("PIP"))
df.columns = header
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(df["iter"], df["mu"].to_numpy())
plt.axhline(y=3.9, linestyle="--", color="k")
plt.title(r"$\mu$")

plt.subplot(2,2,2)
plt.plot(df["iter"], df["alpha"].to_numpy())
plt.axhline(y=10, linestyle="--", color="k")
plt.title(r"$\alpha$")

plt.subplot(2,2,3)
plt.plot(df["iter"], df["sigmaG"].to_numpy())
plt.axhline(y=np.pi**2/(6*10**2), linestyle="--", color="k")
plt.title(r"$\sigma_G$")

plt.subplot(2,2,4)
plt.plot(df["iter"], df["num_markers"].to_numpy())
plt.axhline(y=50, linestyle="--", color="k")
plt.title("Markers included")

plt.tight_layout(pad=1)
```
